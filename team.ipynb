```python
import random
import pandas as pd

lt = []


for i in range(5):
  lt.append(list(range(1, 17)))
for i in range(1, 5):
  random.shuffle(lt[i])


df = pd.DataFrame(lt)
df = df.T #열과 행 전환
df.columns = ['조', '1반', '2반', '3반', '4반']
df = df.set_index('조') #조를 index로 설정


#예외 처리
df.loc[df[df['1반'] == 16].index, '1반'] = '없음'
df.loc[df[df['2반'] == 8].index, '2반'] = '8번, 빠질 수 있음.'
df.loc[df[df['4반'] == 8].index, '4반'] = '없음'
df.loc[df[df['4반'] == 9].index, '4반'] = 17
df
```
